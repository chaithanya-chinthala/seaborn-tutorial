```python
#pairplots-seaborn
#importing packages
import seaborn as sns
import matplotlib.pyplot as plt
#loading dataset using seaborn
df=seaborn.load_dataset('tips')
#df.head()
print(df.to_string())
seaborn.pairplot(df)
plt.show()
#pairplot seaborn:plotting selected variables
df=sns.load_dataset('tips')
selected_vars=['total_bill','tip']
sns.pairplot(df,vars=selected_vars)
plt.show()
#Adding a Hue color to a seaborn pairplot("Hue"=adding an extra dimension here it is color)
df=sns.load_dataset('tips')
seaborn.pairplot(df,hue='size')
plt.show()
#Modifying color palette
sns.pairplot(df, hue='size',palette="husl")
plt.show()
#Diagonal kind of plots("kde"=kernal density estimator)
sns.pairplot(df,diag_kind='kde')
plt.show()
#Adjusting plot kind
sns.pairplot(df,kind='reg')
plt.show()
#controlling the markers
sns.pairplot(df,hue='sex',markers=['o','s'])
plt.show()
#Advanced customization with seaborn pairplot
g=sns.pairplot(df,hue='day')
g.fig.suptitle("pairplot of Tips dataset",y=1.02) #Adding a title
g.set(xticks=[],yticks=[]) #Remove tick labels
plt.show()
```
